# *** WHERE clause***

# Solved Examples on where clause

# select all columns from table where Ethnicity is Asian

```python
sqldf("select * from A where Ethnicity='Asian'")
```

# select data from table where Age is greater than 50

```python
sqldf("select * from A where Age > 50") 
```

# select all columns from table where Ethnicity is not Asian

# ** NOT**

```python
sqldf("select * from A where not Ethnicity ='Asian'")
```

# Select all records where the Cards column has the value 5.

```python
sqldf("select * from A where Cards = 5")
```

# Select all records where the Ethnicity column has the value 'Asian' and the Education column has the value 10.

```python
sqldf("select * from A where Ethnicity ='Asian' AND Education = 10")
```

# Select all records where the Ethnicity column has the value 'Asian' or Caucasian

```python
sqldf("select * from A where Ethnicity ='Asian' or Ethnicity ='Caucasian'")
```

# **** ORDER BY*****

# Syntax
SELECT column1, column2, ...
FROM table_name
ORDER BY column1, column2, ... ASC|DESC;

```python
sqldf("select * from A order by Ethnicity")
```

```python
sqldf("select * from A order by Ethnicity desc")
```

```python
sqldf("select * from A order by Education,Income")
```

```python
sqldf("select *  from A order by Education,Income")
```

```python
sqldf("select Education,Income from A order by Education,Income")
```
